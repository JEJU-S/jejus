```python
추천관광지 = pd.read_csv('추천관광지전처리.csv',encoding='utf-8',index_col=0)
추천관광지
```

```python
추천관광지['SCORE_cal'] = 0
추천관광지['SCORE_cal'] = 추천관광지['SCORE']
추천관광지.head(3)
```

```python
plt.boxplot(추천관광지[['SCORE_cal','SCORE_REVIEW','REVIEW']])
```

```python
sub_df = 추천관광지.loc[(추천관광지['NAME']=='고집돌우럭 제주공항점') | (추천관광지['NAME']=='고집돌우럭 함덕점') | (추천관광지['NAME']=='고집돌우럭 중문점') | (추천관광지['NAME']=='색달식당 중문본점')  ]
sub_df
```

```python
추천관광지.drop(index=[67,1086,1091,2349],inplace=True)
# 추천관광지.drop_duplicates(ignore_index=True,inplace=True)

추천관광지
```

```python
var = ['SCORE_cal','SCORE_REVIEW','REVIEW','IMG_CHECK']
가중치 = [1] * 4 
# weight = [23.5,41.6,20.4,14.5] # AHP CASE 1 
weight = [27.1,34.2,26.3,12.4] # AHP CASE 2 
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
#정규화
norm = 추천관광지.copy()
norm[var] = scaler.fit_transform(norm.loc[:,var])
norm.reset_index(inplace=True)

norm
```

```python
#EDA
plt.boxplot(norm[['SCORE_cal','SCORE_REVIEW','REVIEW']])
```

```python
추천관광지 = norm.copy()

for i in range(0, 4):
    name = var[i]
    추천관광지[name] = 추천관광지[name] * weight[i] 
    
추천관광지.set_index(['index','CATEGORY','SECTION','NAME','SCORE','REGION','LINK','ADDRESS','lat','lng','IMG'], inplace=True)

추천관광지['점수'] = 추천관광지.sum(1) ; 추천관광지.reset_index(inplace=True)

추천관광지['순위'] = 추천관광지['점수'].rank(method='dense', ascending=False).astype(int)

결과 = 추천관광지.sort_values(by='순위') ; 결과.reset_index(drop=True, inplace=True)

display("결과", 결과.head(100))
```

```python
plt.boxplot(결과['점수'])
```

```python
결과['점수'].describe()
```

```python
결과['MODEL'] = 0

결과['MODEL'].loc[( 결과['점수']<23 )] = 'D'
결과['MODEL'].loc[( 결과['점수']>=23 ) & ( 결과['점수']<31 )] = 'C'
결과['MODEL'].loc[( 결과['점수']>=31 ) & ( 결과['점수']<37 )] = 'B'
결과['MODEL'].loc[( 결과['점수']>=37 )] = 'A'


결과
```
