```python
len = df.last_valid_index()
len
```

```python
clist=[]
```

```python
for i in range(len+1):
    driver.get(df['LINK'][i])
    driver.implicitly_wait(7)
    image1 = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/div[1]/div[1]/div[1]/a/span[1]')
    image1_style = image1.get_attribute('style')
    time.sleep(0.1)
    clean1 = image1_style.replace('background-image: url("//','')
    clean2 = re.sub('[-+,#:^$;@*"※~&""ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]','',clean1)
    clist.append(clean2)
    time.sleep(0.1)
    print(df['NAME'][i])
```

```python
df['IMG']=0
for i in range(len+1):
    if clist[i]!='': df['IMG'][i]='http://'+clist[i]
```
