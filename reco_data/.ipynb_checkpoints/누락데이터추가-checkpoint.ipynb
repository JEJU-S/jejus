```python
import pandas as pd

a1 = pd.read_csv('newall.csv',encoding='utf-8')
a2 = pd.read_csv('RecPlace.csv',encoding='utf-8')

df = pd.concat([a1,a2], ignore_index=True)
df
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
df.drop_duplicates(inplace=True,ignore_index=True)
df
```

```python
df.drop_duplicates(['ADDRESS'],inplace=True,ignore_index=True)
df
```

```python
len = df.last_valid_index()
for i in range(len+1):
    if ( df['SCORE'][i]<=3 or df['REVIEW'][i]<=5 or df['SCORE_REVIEW'][i]=='2건' or df['SCORE_REVIEW'][i]=='3건' ):
        df.drop(index=i,inplace=True)
df
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True,ignore_index=True)
df
```

```python
df.to_csv('RecPlace_raw.csv',encoding='utf-8-sig')
```
