```python
import pandas as pd
import time,os,csv,re,json
from IPython.core.display import display ,HTML 
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
dfA = pd.read_csv('sectionA.csv',encoding='utf-8')
dfB = pd.read_csv('sectionB.csv',encoding='utf-8')
dfC = pd.read_csv('sectionC.csv',encoding='utf-8')
dfD = pd.read_csv('sectionD.csv',encoding='utf-8')
dfE = pd.read_csv('sectionE.csv',encoding='utf-8')

df = pd.concat([dfA,dfB,dfC,dfD,dfE], ignore_index=True)

df.tail()
```

```python
df.drop(columns='Unnamed: 0.1',inplace=True)
df.drop(columns='Unnamed: 0',inplace=True)
df.drop_duplicates(inplace=True,ignore_index=True)
```

```python
df
```

```python
df.dtypes
```

```python
len1 = df.last_valid_index()
```

```python
for i in range(len1+1):
    if ( df['SCORE'][i]==0 or df['SCORE_REVIEW'][i]=='0건' or df['SCORE_REVIEW'][i]=='1건' or df['REVIEW'][i]==0.0 or df['REVIEW'][i]==1.0 ):
        df.drop(index=i,inplace=True)
df
```

```python
df.drop_duplicates(inplace=True , ignore_index=True)
df
```

```python
df.to_csv('sections.csv',encoding='utf-8-sig')
```

```python
data = pd.read_csv('sections.csv',encoding='utf-8')
data.tail()
```

```python
data.drop(columns='Unnamed: 0',inplace=True)
data.tail(2)
```

```python
# 중복처리 주소기준
data.drop_duplicates(['ADDRESS'],inplace=True,ignore_index=True)
data
```

```python
len = data.last_valid_index()
```

```python
for i in range(len+1):
    if ( data['SCORE'][i]<=3 or data['REVIEW'][i]<=5 or data['SCORE_REVIEW'][i]=='2건'):
        data.drop(index=i,inplace=True)
data
```

```python
data.dropna(inplace=True)
data
```

```python
data.drop_duplicates(inplace=True,ignore_index=True)
data.tail()
```
