```python
추천관광지 = pd.read_csv('RecPlace_raw.csv',encoding='utf-8',index_col=0)
추천관광지.tail(4)
```

```python
추천관광지['REVIEW'] = 추천관광지['REVIEW'].astype(int)
```

```python
추천관광지['SCORE_REVIEW']=추천관광지['SCORE_REVIEW'].str.replace("건","")
추천관광지['SCORE_REVIEW']=추천관광지['SCORE_REVIEW'].str.replace(pat=r'[^\w]', repl=r'', regex=True)
추천관광지['SCORE_REVIEW']=추천관광지['SCORE_REVIEW'].astype(int)
```

```python
추천관광지.dtypes
```

```python
추천관광지[['part1', 'part2','part3']] = 추천관광지['ADDRESS'].str.partition(sep='(지번)')
```

```python
추천관광지.head(3)
```

```python
추천관광지['part1']=추천관광지['part1'].str.replace('\n','')
추천관광지.head(2)
```

```python
추천관광지['ADDRESS']=추천관광지['part1']
추천관광지.head(2)
```

```python
추천관광지.drop(columns=['part1', 'part2','part3'],inplace=True)
추천관광지.head(2)
```

```python
추천관광지['IMG_CHECK']=0
len = 추천관광지.last_valid_index()
len
```

```python
for i in range(len+1): 
    if 추천관광지['IMG'][i]!= '0' : 추천관광지['IMG_CHECK'][i]=1
추천관광지
```
